```python
from datetime import date
arr_datas_pi_patente = [(date(2022,1,1), None),
                        (date(2021,1,1), None),
                        (date(2020,1,1), (1,date(2022,1,1))),
                        (date(2020,10,1), (1,date(2022,10,1))),
                        (date(2020,12,31), (1,date(2022,12,31))),
                        (date(2017,1,1), (4,date(2022,1,1))),
                        (date(2015,10,1), (6,date(2022,10,1))),
                        (date(2014,12,31), (7,date(2022,12,31)))
                       ]
arr_datas_pi_marca = [(date(2011,1,13)),
                      (date(2021,1,13)),
                      (date(2011,1,13)),
                      (date(2011,1,13)),
                      (date(2011,12,31)),
                       (date(2011,1,13)),
                       (date(2011,1,13)),
                      (date(2001,1,13)),
                      (date(1991,1,13)),
                      (date(1991,12,13)),
                      (date(1989,1,13)),
                      (date(1991,12,13)),
                       ]
arr_concessao_pi_marca = [(date(2011,1,2),None),
                      (date(2021,1,1),None),
                      (date(2011,1,2),None),
                      (date(2011,1,2),(1,date(2022,1,2))),
                      (date(2011,12,31),(1,date(2022,12,31))),
                       (date(2011,1,2),None),
                       (date(2011,1,1),None),
                      (date(2001,1,1),(2,date(2022,1,1))),
                      (date(1991,1,1),(3,date(2022,1,1))),
                      (date(1991,12,31),(3,date(2022,12,31))),
                      (date(1989,1,1),None),
                      (date(1991,12,31),None),
                       ]
```

```python


patente = TipoPropriedadeIntelectual.objects.get(nome=TipoPropriedadeIntelectual.PATENTE)
marca = TipoPropriedadeIntelectual.objects.get(nome=TipoPropriedadeIntelectual.MARCA)
```

```python


for i, (data, _) in enumerate(arr_datas_pi_patente):
    PropriedadeIntelectual.objects.create(titulo=f"PI Patente {i}", codigo=f"TST P {i}", data_deposito=data, tipo_pi=patente, responsavel = cefet_pessoa)


```

```python
for i, (data, _) in enumerate(arr_concessao_pi_marca):
    data_deposito = arr_datas_pi_marca[i]
    pi = PropriedadeIntelectual.objects.get_or_create(titulo=f"PI Marca {i}", codigo=f"TST M {i}", 
                                                              data_deposito=data_deposito, tipo_pi=marca, 
                                                      responsavel = cefet_pessoa)[0]
    
    evento = Evento.objects.create(nome="Concessao Teste :-)",
                                    descricao="Oia, concedida",
                                   pi=pi,
                                  data=data,
                                  tipo_evento=TipoEvento.objects.get(titulo=TipoEvento.OCORRENCIA)
                                  )
    status_concedido = StatusPropriedadeIntelectual.objects.get(nome=StatusPropriedadeIntelectual.CONCEDIDA)
    historico_status_atual = HistoricoStatusPropridadeIntelectual.objects.create(status_pi = status_concedido,
                                                                            pi=pi,
                                                                           evento=evento,
                                                                           data_status=data)
    pi.historico_status_atual = historico_status_atual
```

```python
ano = 2022
arr_pagamentos_por_cobranca = []
arr_pagamentos = []
for cobrancaPrazo in CobrancaPrazo.objects.all().select_related("cobranca","prazo","prazo_extra"):
    for pagamento in cobrancaPrazo.instancia_pagamentos_pis(ano):
        #if pagamento.pi.codigo.find("TST")>=0: 
        arr_pagamentos_por_cobranca.append({"cobranca":cobrancaPrazo.cobranca.nome,
                                                "data_deposito":pagamento.pi.data_deposito,
                                                "data_concessao":pagamento.pi.data_obtencao_status(StatusPropriedadeIntelectual.CONCEDIDA),
                                                "num_pagamento":pagamento.num_pagamento,
                                                 "data_inicio":pagamento.data_inicio_prazo,
                                                 "vencimento":pagamento.data_vencimento,
                                                 "vencimento_extra":pagamento.data_vencimento_extra,

                                               })
        arr_pagamentos.append(pagamento)
data = pd.DataFrame.from_dict(arr_pagamentos_por_cobranca)
data
```
